```python
import pandas as pd
```

```python
countries_df = pd.read_csv('covid-countries.csv')
countries_df
```

```python
covid_df = pd.read_csv('covid-countries-data.csv')
covid_df
```

```python
# Merging covid data and countries data
covid_per_country = countries_df.merge(covid_df, on='location')
covid_per_country
```

```python
num_countries = covid_per_country['location'].shape #Number of countries
print('There are {} countries in the dataset'.format(num_countries))
```

```python
continents = covid_per_country['continent'].unique()
print('These continents {} are shown in the dataset'.format(continents))
```

```python
total_population = covid_per_country['population'].sum()
print('The total population is {}.'.format(int(total_population)))
```

```python
# The number countries in each continent
country_counts_df = covid_per_country.groupby('continent')['location'].count()
country_counts_df
```

```python
# The total population of each continent
continent_populations_df = covid_per_country.groupby('continent')['population'].sum()
continent_populations_df
```

```python
# The overall life expectancy across in the world
weighted_life_expectancy = covid_per_country['life_expectancy']*covid_per_country['population'] # Life expectancy using populations as weights
total_weighted_life_expectancy = weighted_life_expectancy.sum()
total_life_expectancy = covid_per_country['life_expectancy'].sum()
overall_life_expectancy = total_weighted_life_expectancy / total_life_expectancy
print('The overall life expectancy is {}.'.format(overall_life_expectancy))
```

```python
most_popu_df = covid_per_country.sort_values('population', ascending=False)
most_popu_df.head(5)
```

```python
# Overall GDP per country
covid_per_country['gdp'] = covid_per_country['population']*covid_per_country['gdp_per_capita']
covid_per_country
```

```python
# lowest GDP per capita, among the counties with population greater than 100 million
most_population = covid_per_country.population > 10e+7
lowest_gdp_countries = covid_per_country[most_population].sort_values('gdp_per_capita', ascending=True)
lowest_gdp_countries
```

```python
# Test number per million
covid_per_country['tests_per_million'] = covid_per_country['total_tests'] * 1e6 / covid_per_country['population']
# Case number per million
covid_per_country['cases_per_million'] = covid_per_country['total_cases'] * 1e6 / covid_per_country['population']
# Death number per million
covid_per_country['deaths_per_million'] = covid_per_country['total_deaths'] * 1e6 / covid_per_country['population']
```

```python
covid_per_country
```

```python
# The countires that have highest number of tests per million
highest_tests = covid_per_country.sort_values('tests_per_million', ascending=False)
highest_tests[['location','tests_per_million']]
```

```python
# The countires that have highest number of positive per million
highest_cases = covid_per_country.sort_values('cases_per_million', ascending=False)
highest_cases[['location','cases_per_million']]
```

```python
# The countires that have highest number of deaths per million
highest_deaths = covid_per_country.sort_values('deaths_per_million', ascending=False).head(10)
highest_deaths[['location', 'deaths_per_million']]
```
